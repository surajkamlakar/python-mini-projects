```python
#find out no. of courses per subjects
df.head(1)
```

```python
df['subject'].value_counts()
#value_counts() : The value_counts() method returns a Series containing the counts of unique values
```

```python
sns.countplot(df['subject'])
plt.xlabel("Subjects", fontsize = 15)
plt.ylabel("No. of courses per subject", fontsize = 15)
plt.xticks(rotation=60)
plt.show()
```

```python
#For Which Levels, Udemy Courses Providing The Courses
df.columns
```

```python
df['level'].value_counts()
```

```python
#plotting the data
sns.countplot(df['level'])
plt.xlabel("Levels", fontsize = 15)
plt.ylabel("No. of courses per levels", fontsize = 15)
plt.xticks(rotation=60)
plt.show()
```

```python
#Display The Count of Paid and Free Courses
df['is_paid']
```
