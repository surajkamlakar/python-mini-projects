replace income values ['<50K','>50K' with 0 and 1 ]

```python
data.columns
```

```python
data['income'].unique()
```

```python
data['income'].value_counts()
```

```python
sns.countplot('income', data=data)
```

```python
def income_data(sal):
    if sal=='<=50K':
        return 0
    else:
        return 1
```

```python
data['encoded_salary'] = data['income'].apply(income_data)
```

```python
data.head(1)
```

```python
data.replace(to_replace=['<=50K', '>50K'], value=[0,1],inplace= True)
```
