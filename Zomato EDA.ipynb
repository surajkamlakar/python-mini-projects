```python
final_df.Country.value_counts()  ##finding in how many countries Zomato/its stores are available/the number of transactions done 
```

```python
country_names = final_df.Country.value_counts().index ##to find countries and storing the value in variable
```

```python
#to find all ther records for the country names
country_val = final_df.Country.value_counts().values
```

```python
##ploting the pie chart
plt.pie(country_val, labels = country_names)
```

```python
##plotting top 3 countries
plt.pie(country_val[:3], labels = country_names[:3], autopct = '%1.2f%%')
```

## Observations

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text' ]) #we can check the object of that dataframe
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text' ]).size() #we can check the data
```

```python
#convert into dataframe by reset_index
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text' ]).size().reset_index()
```

```python
#giving column name to "0" as Rating Count
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text' ]).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
#saving it in variable
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text' ]).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

#finding relationship of ratings
1. When rating is between 4.5 to 4.9 - Excellent.
2. When rating is between 4.0 to 4.4 - Very Good.
3. When rating is between 3.5 to 3.9 - Good.
4. When rating is between 3.0 to 3.4 - Average.
5. When rating is between 2.5 to 2.9 - Average.
6. When rating is between 2.0 to 2.4 - Poor.

```python
ratings.head()
```

```python
sns.barplot(x = 'Aggregate rating', y= 'Rating Count', data = ratings)
```

# increase the diagram size by using matplotlib

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (14, 6)
sns.barplot(x = 'Aggregate rating', y= 'Rating Count', data = ratings)
```

```python
#get colors
sns.barplot(x = 'Aggregate rating', y= 'Rating Count' , hue= 'Rating color', data = ratings)
```

```python
#mapping the colors
sns.barplot(x = 'Aggregate rating', y= 'Rating Count' , hue= 'Rating color', data = ratings, palette=['blue','white','red','orange','yellow','green','green'])
```

Observation:
    1. Not Rated count is very High
    2. Maximum number of rating are between 2.5 to 3.4

```python
#Count plot is used to plot for categorical varialbe 
sns.countplot(x="Rating color", data=ratings, palette=['blue','red','orange','yellow','green','green'])
```
