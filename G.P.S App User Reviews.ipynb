#it is an object datatype, so inorder to perform any operation we need to change the data type by using 'astype'.

```python
df['Installation'] = df['Installs'].str.replace(',','')
```

```python
df.head(1)
#removed +
```

```python
#now we can convert dtype from obj to int
df['Installation'].astype('int')
```
